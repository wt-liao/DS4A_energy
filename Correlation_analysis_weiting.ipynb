```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

data_path = '/content/drive/MyDrive/DS4A/data'
```

```python
df = pd.read_csv(data_path + "/Cleaned_feature_data/merged_data.csv")
```

```python
df.columns
```

```python
#cluster based on some gdp per person, population, total # & people affected by cc disasters, 2000-2020
# then shade based on above or below average total renewable energy/population
```

```python
variables_included = ['countryname', 'countrycode',\
                      'ClimateDisastersPeopleAffect', 'CCDisastersNumber', \
                      'GDPperPerson', 'population', 'GDP', \
                      'Totalrenewableenergy', 'renew_output', 'renew_consumption', 'access_clean_fuels', 'co2_emissions', \
                      'AllDisastersNumber', 'AllDisastersPeopleAffected', 'AlldisastersTotalDamages']
df_clean = df[variables_included]
```

```python
corr = df_clean.corr()

fig, ax = plt.subplots(figsize=(7,7)) 
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
)

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);

plt.savefig(data_path+"/temp.png", bbox_inches = "tight")
```
