```python
all_months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', \
              'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
data_out = np.zeros([len(all_country), 4], dtype=object) # ave sunshine hours, std, # data points, # pts with nan
for i, country in enumerate(all_country):
    hrs = np.array([])

    for d_country in df['Country or Territory'].unique():
        if country in d_country:
            for month in all_months:
                data = df.loc[df['Country or Territory'] == d_country][month]
                hrs = np.append(hrs, np.array(data))
    
    data_out[i, :] = [np.nanmean(hrs), np.nanstd(hrs), len(hrs), np.isnan(hrs).sum()]
                
```

```python
data_out
```

```python
df_out = pd.DataFrame(data_out, \
                      columns = ["ave sunshine hours per month", "std", "total data points", "bad data points"], 
                      index = all_country)
```
